```python
import numpy as np
import matplotlib.pyplot as plt
import heartpy as hp
```

```python
data = hp.get_data('capture01_250Hz.txt')
Fs = 250 
time = np.arange(len(data))/Fs
```

```python
plt.figure(figsize = (12, 4))
plt.plot(time, data)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (uint16)')
plt.show()
```

```python
np.random.seed(2852)
t0 = np.random.randint(30, 200)
print(f'Segment starts from: {t0} s')
i = (t0 < time) & (time < t0 + 60)
data2 = data[i]
time2 = time[i]

plt.figure(figsize = (12, 4))
plt.plot(time2, data2)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (uint16)')
plt.show()
```

```python
plt.figure(figsize = (12, 4))
plt.plot(time2, data2)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (uint16)')
plt.xlim(t0, t0 + 6)
plt.show()
```
